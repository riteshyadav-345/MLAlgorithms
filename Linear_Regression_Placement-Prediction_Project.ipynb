```python
X=df.iloc[:,0:1] # Selects only 'cgpa' column as feature
y=df.iloc[:,-1] # Selects 'package' as target
```

```python
X  #input
```

```python
y   #output
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
```

```python
clf.fit(X_train,y_train)    #Model Training
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(df['cgpa'], clf.predict(scaler.transform(X)), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.title('CGPA vs Package with Regression Line')
plt.show()
```
